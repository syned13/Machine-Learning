```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = datasets.load_breast_cancer()
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df['Target'] = data.target
df
```

```python
df.corr()
```

```python
for column in df:
    df[column].hist()
    plt.title(column)
    plt.show()
```

```python
# compactness tells us how irregular the boundary of the cell nucleus is
plt.scatter(df['mean compactness'], df['Target'], alpha=0.4)
plt.xlabel('mean compactness')
plt.ylabel('Target')
```

```python
for column in df:
    if column == 'Target':
        continue
    
    plt.scatter(df[column], df['Target'])
    plt.xlabel(column)
    plt.ylabel('Target')
    plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('Target', axis=1), df['Target'])
scaler = StandardScaler()
X_train_normalized = scaler.fit_transform(X_train)
model = KNeighborsClassifier(n_neighbors=6).fit(X_train_normalized, y_train)
```

```python
model.score(scaler.fit_transform(scaler.fit_transform(X_test)),y_test)
```

```python
y_predicted = model.predict(scaler.fit_transform(X_test))
```

```python
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(model,scaler.fit_transform(X_test), y_test)
```

```python
from sklearn.metrics import precision_score
ks = list(range(3,101))
scores = []
for k in ks:
    model = KNeighborsClassifier(n_neighbors=k).fit(X_train_normalized, y_train)
    predicted = model.predict(scaler.fit_transform(X_test))
    scores.append(precision_score(y_test, predicted))
#     scores.append(model.score(scaler.fit_transform(X_test),y_test))
```

```python
plt.plot(ks, scores)
plt.xlabel('K')
plt.ylabel('Score')
```
